### Project-Identifying and Recommending Best Restaurants

* Importing ,Understanding and Inspecting Data

```python
# importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# importing dataset using pandas
df=pd.read_excel('data.xlsx')
df.head(10)  # first 10 records of dataset
```

```python
# importing country_code dataset
Country_Code_df=pd.read_excel('Country_Code.xlsx')
```

```python
#now merging dataframe using pandas merge function
df1=pd.merge(df,Country_Code_df,on='Country Code',how='left')
df1.head(10) # first 10 records of dataframe
```

```python
# Checking all columns in dataset
df1.columns
```

```python
# renaming column name because some column name are having space
df1.columns=df1.columns.str.replace(' ','_')
```

```python
# Checking data type of all column
df1.info()
```

```python
# checking null values in each column of the dataframe
df1.isna().sum()
```

* We can see above here two columns of dataframe is having nan value.

```python
#checking null value in Restaurant_Name
df1[df1['Restaurant_Name'].isna()]
```

```python
#Checking null value in Cuisines name
df1[df1['Cuisines'].isna()]
```

```python
# dropping null value in Restaurant_Name
df1.dropna(subset=['Restaurant_Name'],inplace=True)
# dropping null value in Cuisines
df1.dropna(subset=['Cuisines'],inplace=True)
```

```python
# Checking again null value in each column name using heatmap
sns.heatmap(df1.isna(),cmap='viridis')
plt.show()
```

```python
# Checking duplicate row in dataframe
df1.duplicated().sum()
```

#### We can see here there is no duplicate row in dataframe.

```python
# removing '#'from Restaurant_Name
df1['Restaurant_Name']=df1['Restaurant_Name'].str.replace('#','')
```

#### Writing cleaned dataframe

```python
# writing dataframe into csv format after inspecting data
df1.to_csv('cleaned_data.csv',index=False)
```

```python
#Copying full dataframe
df2=df1.copy()
df2.head(10) # printing first 10 records of dataframe
```

```python
# converting categorical column into numerical using one hot encoding for finding ratio b/w them
df2=pd.get_dummies(df1,columns=['Has_Table_booking','Has_Online_delivery'],drop_first=True)
df2.head(10)
```

#### Here we can see that 0 indicating No and 1 Indicating Yes in one hot encoding

```python
# finding ratio between restaurant that allow table booking and that do not allow table booking
table_booking_yes=df2[df2['Has_Table_booking_Yes']==1]['Restaurant_ID'].count()
table_booking_no=df2[df2['Has_Table_booking_Yes']==0]['Restaurant_ID'].count()
print('The total number of Restaurant that allow table booking is',table_booking_yes)
print('The total number of Restaurant that do not allow table booking is',table_booking_no)
```

```python
print('The ratio b/w restaurant that allow table booking vs those do not allow table booking is',table_booking_yes/table_booking_no)
```

```python
# plotting pie chart using matplotlib
mydata=[table_booking_yes,table_booking_no]
label=['Table Booking Yes','Table Booking No']
colors=['Yellow','Cyan']
myexplode=[0.2,0] # explode is used for wedges standout
plt.pie(x=mydata,labels=label,autopct='%.2f%%',startangle=90,explode=myexplode,shadow=True,colors=colors,)
plt.title('Table Booking Yes Vs Table Booking No ',weight='bold',color='blue')
plt.show()
```

```python
# finding percentage restaurant that providing online delivery
online_delivery_yes=df2[df2['Has_Online_delivery_Yes']==1]['Restaurant_ID'].count()
online_delivery_no=df2[df2['Has_Online_delivery_Yes']==0]['Restaurant_ID'].count()
sum_of_both=online_delivery_yes+online_delivery_no
percentage=online_delivery_yes*100/sum_of_both
print('Percentage of Restaurant that providing online delivery is {} %'.format(round(percentage)))
```

```python
# plotting pie chart using matplotlib
mydata=[online_delivery_yes,online_delivery_no]
label=['Online Delivery Yes','Online Delivery No']
myexplode=[0.1,0] # explode is used for wedges standout
plt.pie(x=mydata,labels=label,autopct='%.2f%%',startangle=90,explode=myexplode,shadow=True)
plt.title('Online Delivery Yes Vs Online Delivery No',weight='bold',color='blue')
plt.show()
```

```python
# finding difference in number of votes
online_delivery_yes=df2[df2['Has_Online_delivery_Yes']==1]['Votes'].sum()
online_delivery_no=df2[df2['Has_Online_delivery_Yes']==0]['Votes'].sum()
print('Total votes for having online delivery is',online_delivery_yes)
print('Total votes for restaurant that do not deliver online is',online_delivery_no)
```

```python
print('Difference in number of votes is',abs(online_delivery_yes-online_delivery_no)) # abs is used for absolute value
```
